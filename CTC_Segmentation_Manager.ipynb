```python
import pandas as pd
import librosa
import json

def dump2json(data, manifest_path):
    with open(manifest_path, 'w') as fout:
        for metadata in data:
            json.dump(metadata, fout)
            fout.write('\n')
            
def process_record(audio_root, x):
    audio_filepath = audio_root + x.wav_filename.split('/')[-1]
    return {'audio_filepath': audio_filepath, 
            'duration': librosa.core.get_duration(filename=audio_filepath),
            'text': x['transcript']
           }

df_train = pd.read_csv('./dataset/ted-train.csv', sep=',')
df_test = pd.read_csv('./dataset/ted-test.csv', sep=',')
df_dev = pd.read_csv('./dataset/ted-dev.csv', sep=',')


audio_root_dev = '/home/wesam/dataset/TEDLIUM_release2/dev/wav/'
audio_root_train = '/home/wesam/dataset/TEDLIUM_release2/train/wav/'
audio_root_test = '/home/wesam/dataset/TEDLIUM_release2/test/wav/'


dev = list(df_dev.apply(lambda x: process_record(audio_root_dev, x), axis=1))
train = list(df_train.apply(lambda x: process_record(audio_root_train, x), axis=1))
test = list(df_test.apply(lambda x: process_record(audio_root_test, x), axis=1))

dump2json(data = dev, manifest_path = '/home/wesam/dataset/TEDLIUM_release2/ted_dev_nemo.json')
dump2json(data = train, manifest_path = '/home/wesam/dataset/TEDLIUM_release2/ted_train_nemo.json')
dump2json(data = test, manifest_path = '/home/wesam/dataset/TEDLIUM_release2/ted_test_nemo.json')

```
